```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# from math import * 
```

```python
data = pd.read_csv('sample_database')
```

```python
column_titles = [i for i in data.columns]
columns_numerical = column_titles[:8]
columns_signal = [f'signal_{i}' for i in range(1, 801)]
columns_background = [f'background_{i}' for i in range(1, 801)]
columns_sample = [f'sample_{i}' for i in range(1, 801)]
```

#### Functions

```python
def narrow_down_dataframe(df, beam_low, beam_high, l_low, l_high):
    beam=df.loc[df['beam_div'] > beam_low]
    beam=beam.loc[df['beam_div'] < beam_high]
    beam_lambda=beam.loc[beam['wavelength'] > l_low]
    beam_lambda=beam_lambda.loc[beam_lambda['wavelength'] < l_high]
    return beam_lambda[columns_numerical], beam_lambda[columns_signal], beam_lambda[columns_background], beam_lambda[columns_sample]
```

```python
fig, ax = plt.subplots(figsize=(12, 12))

colors = sns.color_palette('Paired')[0:12]

material_count = data['material_verbose'].value_counts()
materials = material_count.index.tolist()
counts = material_count.tolist()
total=sum(counts)

wedges, texts, autotexts = ax.pie(counts, labels=materials, autopct='%1.1f%%', startangle=90, colors=colors, textprops={'fontsize': 20}, pctdistance=0.8)
ax.set_title("Sample materials", fontsize=32)

autopct_labels = [f' {val*100/total:.1f} %' for val in counts]
autopct_with_box = [f'{val}  ' for val in autopct_labels]

for i, autotext in enumerate(autotexts):
    autotext.set_text(autopct_with_box[i])

plt.show()
```

```python
corr_matrix = data[columns_numerical].corr(method='pearson')
fig, ax = plt.subplots(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
parameters, total, background, sample = narrow_down_dataframe(I2, beam_low=20, beam_high=40, l_low=1.1, l_high=5)
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(15, 15),sharex=True)

title=30
ticks=25
legend=25
label=30

n=0
x=np.linspace(10,170,len(total.iloc[n].T))
y_axis_range = (-50, max(total.iloc[n])*1.1)

ax1.plot(x, total.iloc[n].T, color='navy', linewidth=7, label='Full Signal')
ax1.set_title(f'{parameters.iloc[n][-1]}',fontsize=title)
ax1.set_ylim(y_axis_range)

ax2.plot(x, background.iloc[n].T, color='forestgreen', linewidth=7, label='Background', alpha=0.75)
ax2.plot(x, sample.iloc[n].T, color='firebrick', linewidth=7, label='Sample', alpha=0.75)
ax2.set_title('Signal Components',fontsize=title)
ax2.set_ylim(y_axis_range)

ax1.legend(fontsize=legend)
ax1.tick_params(axis='both', labelsize=ticks)
ax1.set_ylabel('Intensity [n/s/bin]',fontsize=label)

ax2.legend(fontsize=legend)
ax2.set_ylabel('Intensity [n/s/bin]',fontsize=label)
ax2.tick_params(axis='both', labelsize=ticks)
ax2.set_xlabel(r'2$\theta$ [$^o$]',fontsize=label)

plt.tight_layout(h_pad=2)
plt.show()
```
